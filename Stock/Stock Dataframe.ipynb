```python
import numpy as np
import pandas as pd
from datetime import *
```

```python

Country_Stockmarket_Dict1 = {"USA":"DJIA"} #Sample
Country_Stockmarket_Dict = {
"Austria":"ATX Index",
"Belgium":"Bel-20",
"Denmark":"OMX Copenhagen",
"Finland":"OMX Helsinki",
"France":"CAC 40",
"Germany":"DAX",
"Greece":"Athex Composite",
"Italy":"FTSE MIB",
"Netherlands":"AEX",
"Norway":"OBX Index",
"Portugal":"PSI 20",
"Russia":"RTS Index",
"SouthAfrica":"FTSE JSE Africa All Share",#
"Spain":"IBEX 35",
"Sweden":"OMX Stockholm 30",
"Switzerland":"Swiss Market",
"Turkey":"BIST 100",
"UnitedKingdom":"FTSE 100",#
"Australia":"S&P ASX",
"China":"Shanghai Composite Index",
"HongKong_China":"Hang Seng",#
"India":"S&P BSE Sensex",
"Indonesia":"JSX Index",
"Japan":"Nikkei 225",
"Malaysia":"FTSE Bursa Malaysia KLCI",
"NewZealand":"S&P NZX 50",#
"Philippines":"PSEi Index",
"KoreaSouth":"KOSPI",#
"Singapore":"Straits Times",
"Thailand":"SET",
"Brazil":"Bovespa",
"Canada":"S&P TSX Composite",
"Mexico":"IPC",    
"UnitedStates":"DJIA",#
}
#set date
Date_Template = pd.DataFrame({"Date": pd.date_range('2020-02-01', datetime.today())})
#print(Date_Template.head())
#print(Date_Template.tail())
Stock = Date_Template

for Country, Market in Country_Stockmarket_Dict.items():
    #print(Country+"_"+Market+".csv")
    Dummy = pd.read_csv(Country+"_"+Market+".csv",skipinitialspace=True).sort_values("Date")
    Dummy[Country] = Dummy["Close"]
    Dummy["Date"] = pd.to_datetime(Dummy["Date"], infer_datetime_format=True)
    Dummy = Dummy.drop(["Open","High","Low","Close"],axis=1)
    #print(Dummy.head())
    Stock = Stock.merge(Dummy, on="Date",how="left")
print(Stock.tail())

```
